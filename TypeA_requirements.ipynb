# Type-A suspension requirements

by K. Okutomi, 2018.11.07

*Note: Translation of this document to English is not scheduled. Please wait and see my PhD thesis.*

著者がD論を書くにあたり、Type-A suspensionの要求値をまとめ直したもの。過去に要求値を検討された時と2018年現在ではいくつかのパラメータが変わっていたり一部未確定だった仕様が確定したりしていると思われるので、特にそれらの更新されたものに注意して要求値を再検討した。


## 干渉計の制御フェイズ

まず初めに<a href="#SekiguchiPhD">[3] 関口D論</a>では以下の3つの制御フェイズを定義しており、VISでは各制御フェイズごとに要求値を考えている。

1. **Calm-down phase**
 - 干渉計信号(REFL, WFS, etc...)が取れないくらいsuspensionの振動が励起されている段階。Noiseは気にせずとにかくdampingを行う。
 
1. **Lock acquisition phase**
 - 干渉計信号が復帰し、鏡の共振引き込みを行う段階。できるだけ早くロックを達成する。
 
1. **Observation phase**
 - 観測運転状態。低雑音・高安定性な干渉計制御を目的とする。

実際にはGuardianによる状態遷移を行うので、上記の制御フェイズの枠組みは厳密には正しくないかもしれない。
しかし以下の要求値の議論においては上記の制御フェイズの枠組みを採用する。

## 各要求値の導出・考察

閲覧者（特に干渉計の人たち）に確認したい重要度順に書いてある。

<a id="RMSvelL"></a>
### RMS velocity (L) for lock acquisition

L方向のRMS velocityの要求値は2つの制約に基づいた計算方法がある。

1. 鏡がPDH線形域を横切る時間スケールが制御帯域の時間幅と比べて十分遅いこと
1. 鏡がPDH線形域を横切る間にアクチュエータのできる仕事が鏡の運動エネルギーよりも大きいこと（あるいは鏡の運動量について同等の制約）

現在設定されているのは前者の制御帯域の制約からくる要求値である。<a href="#SekiguchiPhD">[3] 関口D論</a>によれば、鏡のPDH線形域への進入速度 $v_\mathrm{in}$が満たすべき条件は、

\begin{align}
v_\mathrm{in} \lesssim \omega_b \Delta L_\mathrm{lin} = \frac{\omega_b \lambda}{2\mathcal{F}}
\end{align}

となる。ここで$\omega_b$は制御帯域、$\Delta L_\mathrm{lin}$はPDH線形域の幅、$\lambda$はレーザー波長（green: 532 nm）、$\mathcal{F}$は注目しているレーザー波長での共振器のフィネスである。<a href="#SekiguchiPhD">[3] 関口D論 p.90</a>では制御帯域の値が鏡アクチュエータだったり周波数アクチュエータだったり、またレーザー波長がIRだったりgreenだったりで、鏡速度の要求値の計算に使う値が安定していない。そこで鏡アクチュエータによる制御帯域を50 Hz、green laserに対する共振器のフィネスを50として計算し直すと、**Type-A suspensionにおいてgreen lockに要求される鏡の進入速度 $v_\mathrm{in}$は$\lesssim\, 1.6\;\mathrm{um/s}$となる**。

以上より、** RMS velocityに対する要求値 **は、「鏡速度が正規分布に従うと仮定し、$v_\mathrm{in}$よりも小さい速度で鏡がPDH線形域に進入する確率が50 %以上である」という要求を設けた場合、

\begin{align}
v_\mathrm{rms} \lesssim 0.72\, v_\mathrm{in} = 1.1\;\mathrm{[um/s]}
\end{align}

と求められる。

もう1つのアクチュエータレンジの制約について言及している新しい文献としては<a href="#Michimura2017">[4]</a>がある。ここでは他のtypeのsuspensionについて鏡アクチュエータのmax forceからRMS velocityに対する要求値を計算しているが、Type-Aについては周波数アクチュエータによるgreen lockを行うというstrategyのためRMS velocityの要求値を定めていない。（おそらく当時AOMの仕様が未確定だったため？）

> For the Type-A suspension, lock acquisition is done with the laser frequency
actuator of the arm-length stabilization system, with the feedback point then gradually
changed from the frequency actuator to the mirror actuator. Therefore, the discussion
above cannot be applied.

<a href="#Michimura2018">[7] ALS schematics</a>を見る限り、片腕のgreen lockは以下のような手順で行われるものと奥富は解釈している。

** Green lock procedure （仮）**

1. 周波数アクチュエータ(80 MHz VCO + AOM)を使ってgreen laserをarm cavityにロック
2. feedback先を徐々に周波数アクチュエータから鏡アクチュエータに切り替えて低周波帯の制御をhand over (< 50 Hz?)
3. AUXの2次光？(1064 nm)とPSLのbeatを取って位相同期
4. green laser光源周波数ががPSL周波数に追従するようになる -> 鏡アクチュエータ制御帯域ではarm cavity長がPSL周波数に追従する
5. greenとPSLのoffsetを調整して鏡（cavity長）をPSLのPDH線形域に引き込み完了

この2.において鏡アクチュエータに制御をhand overする際、Type-A suspensionの階層制御およびALS制御の詳細設計が未確定なため、鏡アクチュエータのmax powerからRMS velocityの要求値を求めるのが現時点では難しいってことなのかな…？（理解が曖昧）
このあたり、最新の状態においてもし周波数アクチュエータの仕様が定まっていたりすれば、アクチュエータレンジの制約からRMS velocityの要求値を計算できたりするのかもしれない？

<a id="RMSdispL"></a>
### RMS displacement (L) in observation


<a href="#SekiguchiPhD">[3] 関口D論 p.92</a>ではTMアクチュエータで出力可能なmax forceからRMS displacementの要求値を定めているが、それと同時にType-Aの場合には階層制御によってアクチュエータパワーの強い上段にFB先をhand offしていく必要があると述べている。

> Note that the requirement values are set with the assumption that the interferometer
lock is maintained only by the actuators on the mirrors. On the other hand, one can
distribute the low frequency component of the interferometer control signals to the upperstage
actuators, which are allowed to have larger actuation range. This scheme is called
hierarchical control and would be required especially for the type-A system, in which
the applicable force by the mirror actuators is extremely weak.

Type-Aの場合にはL方向のRMS displacementの要求値を与えるのではなく、上段まで含めた階層制御においてfeedback signalがアクチュエータの最大出力(DAC-limited?)を上回らない制御系となることを要求するのが妥当ではないかと考えられる。そこで**今回（私のD論）では観測フェイズにおけるType-AでのRMS displacement (L)の要求値を除外しておく**。


<a id="RMSanglePY"></a>
### RMS angle (P, Y)

アラインメント角度変動のRMSに対する要求値は、3 km先のbeam spot位置変動が1 mm以下になるよう設定されている。この1 mmの根拠については2012年の段階では明確な基準は特にないようだった<a href="#Michimura2012">[1]</a>。しかし、lock acquisition以降の制御フェイズにおいてはP, Y角度を見るOpLevのローカル制御がWFS制御への前段階であると考えると、P, Y方向のRMS angleの要求値はWFSの線形域の大きさ程度となるのが妥当ではないかと思われる。

2012年の時点では、このbeam spot揺れが1 mm以下という基準のもと、Optickleを用いたBMS行列の計算によって各鏡ごとのRMS angleの要求値を求めている。しかし2017年に感度曲線がupdateされたため、榎本くんが各鏡の変位雑音を計算し直してくれたように<a href="#Enomoto2018">[9]</a>、**鏡のアラインメント角度揺れについてもBMS行列を更新して要求値を計算し直す必要性があるかもしれない**。

また現在進行中のbKAGRA phase2 X-arm commissioningにおいては、RMS angle (P, Y)に対する要求値はoptical gainの変動が5 %以下となるよう< 880 nradと設定されている。これは従来の要求値である200 nradよりも大きく、あくまでbKAGRA phase-2 commissioningにおいて設定された値であると考えられるので、現状Type-A suspensionの角度揺れRMSの要求値としては0.2 uradを採用するのが妥当と思われる（値の更新を前提としないのであれば）。

なお、ここで言及しているのはOpLevを用いた鏡角度のローカル制御までであり、WFSを用いたグローバル制御？に対する要求値を考えているわけではないことに注意。そのため、lock acquisition phase および observation phase において同じ要求値が採用されている。

<a id="RMSdispTV"></a>
### RMS displacement (T, V)

T, V方向のRMS displacementについては全ての制御フェイズにおいて、<a href="#Aso2013">[2] p.13</a>でWFS shot noise requirementを計算する際のmiscenteringの見積もりがそのまま現在の要求値となっている。

> Assuming rms angular fluctuations
of the mirrors to be less than $10^{-8}$ rad, rms beam-spot
motions on the mirrors are estimated to be smaller than
0.1 mm. Therefore, we use $d_m$ = 0.1 mm for the test
masses in the following calculations.

またlong-term driftについても同様の値が要求値として設定されている。0.1 mmという値はかなり大きいので、地面が十分静かであればsuspensionの制御においてそこまで頑張らなくても達成可能な要求であると予想されるが、少なくともcalm-down phaseにおいては本要求値の設定は必要ないのではないかと思われる。

<a id="DCdriftL"></a>
### DC drift (L)

<a href="#SekiguchiPhD">[3] 関口D論 p.93</a>ではRMS displacementと同様の値が長期安定性に関する要求値に設定されている。しかしType-Aに限らず鏡のDC driftはIPなどの上段アクチュエータ or モータで補正することになるので、<a href="#RMSdispL">sec. 3.2</a>のRMS displacementの要求値をそのまま流用するのは適当でないと考えられる。

<a id="DCdriftPY"></a>
### DC drift (P, Y)

<a href="#Michimura2012">[1]</a>においてOpLevのpointing driftに対する要求値が< 0.4 urad/hと提案されており、<a href="#SekiguchiPhD">[3] 関口D論</a>においてもその値がDC drift (P, Y)の要求値として採用されている。OpLevの仕様が変更になっていなければ、この値は現在でも有効な値であると考えられる。

<a id="decaytime"></a>
### Modal $1/e$ decay time

典型的な値として1分以内という減衰時間の要求値が与えられている。以前和泉さんと相談した際には「まぁ1分以内にdampできれば大丈夫なんじゃない？」との回答を頂いており、経験的な値ではあるがsuspensionローカルだけでなく干渉計側から見ても妥当な値なのだろうと認識している。しかし自由度によっては1分よりも短くしたいという要求もあったりするので<a href="#Shoda2018">[6]</a>、必要に応じて再検討していく方針で良いと思われる。
