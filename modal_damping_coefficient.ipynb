# Modal damping coefficient

## 不減衰系におけるモード分解

不減衰系におけるモード解析では、

\begin{align}
(-\omega^2_i\boldsymbol{M} + \boldsymbol{K})\,\boldsymbol{\phi}_i = \boldsymbol{0}
\end{align}

という固有値問題を解くことにより、系の固有角周波数$\omega_i$と固有ベクトル$\boldsymbol{\phi}_i$を得る。
ここで$\boldsymbol{M},\,\boldsymbol{K}$はそれぞれ物理座標系$\boldsymbol{x}$における質量行列と剛性行列である。

上で求めた固有ベクトルを列ベクトルとして横に並べた行列を固有モード行列$\boldsymbol{\Phi}$と呼び、
モード座標系$\boldsymbol{\eta}$と物理座標系$\boldsymbol{x}$の座標変換を与える。

\begin{align}
\boldsymbol{x} &= \boldsymbol{\Phi\,\eta} \\
\boldsymbol{\Phi} &\equiv [\boldsymbol{\phi}_1\;\boldsymbol{\phi}_2\;\dots\;\boldsymbol{\phi}_n]
\end{align}

$r$番目の固有ベクトル$\boldsymbol{\phi}_r$で物理座標系における質量行列および剛性行列の2次形式（固有ベクトルとその転置で挟み込む）を取ると、それぞれ$r$次モードのモード質量$m_r$とモード剛性$k_r$を得る。

\begin{align}
\boldsymbol{\phi}_r^\top \boldsymbol{M} \boldsymbol{\phi}_r &= m_r \\
\boldsymbol{\phi}_r^\top \boldsymbol{K} \boldsymbol{\phi}_r &= k_r \\
\end{align}

## 減衰系におけるモード分解

いま、2つのタイプの減衰過程を考える。

1. 外部から系に作用する散逸過程（周囲の流体など）
1. 系の内部に由来する散逸過程（材質の内部エネルギー損失、系の結合部や接触部における構造減衰など）

具体的な行列要素の表式を分かりやすくするため、マスAとマスBからなる2自由度系を考える。

\begin{align}
\boldsymbol{M}\ddot{\boldsymbol{x}} + \boldsymbol{C}\dot{\boldsymbol{x}} + \boldsymbol{Kx} = \boldsymbol{0}
\end{align}

この系のモード分解は、運動方程式に特殊解 $x_a = X_a \,e^{\lambda t}$, $x_b = X_b \,e^{\lambda t}$を代入して得られる方程式

\begin{align}
(\lambda^2 \boldsymbol{M} + \lambda\boldsymbol{C} + \boldsymbol{K})
\left[\begin{matrix}
X_a \\ X_b
\end{matrix}\right] = 0
\end{align}

で $X_a = X_b = 0$ 以外の解を求める固有値問題となる。

減衰行列$\boldsymbol{C}$を減衰過程ごとに分けて表現

\begin{align}
\left[\begin{matrix}
c_{ma} + c_{ka} + c_{kb} & -c_{kb} \\
-c_{kb} & c_{mb} + c_{kb}
\end{matrix}\right] = 
\left[\begin{matrix}
c_{ma} & 0 \\
0 & c_{mb}
\end{matrix}\right] +
\left[\begin{matrix}
c_{ka} + c_{kb} & -c_{kb} \\
-c_{kb} & c_{kb}
\end{matrix}\right]
\end{align}

この右辺は第1項が質量行列と、第2項が剛性行列と、それぞれ見かけの形が似ている。そこで**減衰行列は質量行列と剛性行列の線形結合で書ける**という仮定をおき、その比例定数を$\alpha,\,\beta$とすると

\begin{align}
\left[\begin{matrix}
c_{ma} + c_{ka} + c_{kb} & -c_{kb} \\
-c_{kb} & c_{mb} + c_{kb}
\end{matrix}\right] = 
\alpha \left[\begin{matrix}
m_a & 0 \\
0 & m_b
\end{matrix}\right] +
\beta \left[\begin{matrix}
k_a + k_b & -k_b \\
-k_b & k_b
\end{matrix}\right]
\end{align}

\begin{align}
\therefore \quad \boldsymbol{C} = \alpha\boldsymbol{M} + \beta\boldsymbol{K}
\end{align}

\begin{align}
\left[(\lambda^2 + \alpha\lambda)\, \boldsymbol{M} + (\beta\lambda + 1)\,\boldsymbol{K}\right]
\left[\begin{matrix}
X_a \\ X_b
\end{matrix}\right] = 0
\end{align}
